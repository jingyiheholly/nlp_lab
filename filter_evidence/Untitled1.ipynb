```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as pch
plt.rcParams["figure.figsize"] = 10,10
fig, ax = plt.subplots()
start_x=0;start_y=0 #start point
terminal_x=0;terminal_y=140 #terminal point,cm
ax.plot(start_x,start_y,marker='X',markersize='20',markeredgewidth=0.01,color='k')
ax.text(start_x-20, start_y-10,"start point",fontsize=14,color="k",weight='bold')
ax.text(start_x-15, start_y+10,"(0,0)",fontsize=14,color="k",weight='bold')
ax.plot(terminal_x,terminal_y,marker='X',markersize='20',markeredgewidth=0.01,color='b')
ax.text(terminal_x-12, terminal_y-10,"terminal point",fontsize=14,color="b",weight='bold')
ax.text(terminal_x-22, terminal_y+10,"(0,140)",fontsize=14,color="b",weight='bold')
x_1=150;y_1=0
x_2=150;y_2=140
ax.plot(x_1,y_1,marker='X',markersize='10',markeredgewidth=0.01,color='k')
ax.text(x_1-12,y_1-10,"point 1",fontsize=14,color="k",weight='bold')
ax.text(x_1+10,y_1+10,"(150,0)",fontsize=14,color="k",weight='bold')
ax.plot(x_2,y_2,marker='X',markersize='10',markeredgewidth=0.01,color='b')
ax.text(x_2-12,y_2-10,"point 2",fontsize=14,color="b",weight='bold')
ax.text(x_2+10,y_2+10,"(150,140)",fontsize=14,color="b",weight='bold')
plt.arrow(start_x,start_y,x_1-start_x,y_1-start_y,width=0.5,linestyle=':',length_includes_head=1,head_length=5,head_width=3,color='k')
plt.arrow(x_1,y_1,x_2-x_1,y_2-y_1,width=0.5,linestyle=':',length_includes_head=1,head_length=5,head_width=3,color='k')
plt.arrow(x_2,y_2,terminal_x-x_2,terminal_y-y_2,width=0.5,linestyle=':',length_includes_head=1,head_length=5,head_width=3,color='k')
obstacle_1=pch.Rectangle((135,50),width=45,height=35,color='c',alpha=0.6)
obstacle_2=pch.Rectangle((80,120),width=20,height=60,color='r',alpha=0.6)
ax.add_patch(obstacle_1)
ax.add_patch(obstacle_2)
obstacle_1_v=pch.Rectangle((125,40),width=65,height=55,fill=0,linewidth=2.5,linestyle=':',color='c',alpha=1)
obstacle_2_v=pch.Rectangle((70,110),width=40,height=80,fill=0,linewidth=2.5,linestyle=':',color='r',alpha=1)
ax.add_patch(obstacle_1_v)
ax.add_patch(obstacle_2_v)
point_x=[125,190,190,125,70,110,110,70]
point_y=[40,40,95,95,110,110,190,190]
plt.plot(point_x,point_y,'o')
ax.axis([-50,200,-50,200])
ax.set_aspect("equal")
# plt.grid(b=None, which='major', axis='both') #in new matplotlib version ，instead 'b' by 'visible'
plt.grid(visible=None, which='major', axis='both')
x = range(-50,200,10)
y= range(-50,200,10)
plt.xticks(x)
plt.yticks(y)
#plt.show()
p1=[x_1,y_1]
p2=[125,40]
p3=[190,40]
p4=[125,95]
p5=[190,95]
p6=[x_2,y_2]
p7=[110,110]
p8=[110,190]
p9=[70,110]
p10=[70,190]
p11=[terminal_x,terminal_y]
def distance(P1,P2):
    distance=((P1[0]-P2[0])**2+(P1[1]-P2[1])**2)**0.5
    return distance
inf = 10000 
mgraph =[[0,distance(p1,p2),distance(p1,p3),inf,inf,inf,inf,inf,inf,inf,inf], [inf,0,distance(p2,p3),distance(p2,p4),inf,inf,inf,inf,inf,inf,inf], [inf,distance(p2,p3),0,inf,distance(p3,p5),inf,inf,inf,inf,inf,inf], [inf,inf,inf,0,distance(p4,p5),distance(p4,p6),inf,inf,inf,inf,inf], [inf,inf,inf,distance(p4,p5),0,distance(p5,p6),inf,inf,inf,inf,inf], [inf,inf,inf,inf,inf,0,distance(p6,p7),distance(p6,p8),inf,inf,inf], [inf,inf,inf,inf,inf,inf,0,distance(p7,p8),distance(p7,p9),inf,inf], [inf,inf,inf,inf,inf,inf,distance(p7,p8),0,inf,distance(p8,p10),inf], [inf,inf,inf,inf,inf,inf,inf,inf,0,distance(p9,p10),distance(p9,p11)], [inf,inf,inf,inf,inf,inf,inf,inf,distance(p9,p10),0,distance(p10,p11)], [inf,inf,inf,inf,inf,inf,inf,inf,inf,inf,0]] 
def startwith(start: int, mgraph: list) -> list: 
    passed = [start] 
    nopass = [x for x in range(len(mgraph)) if x != start] 
    dis = mgraph[start] 
    dict_ = {} 
    for i in range(len(dis)): 
        if dis[i] != inf: 
            dict_[str(i)] = [start] 
    while len(nopass): 
        idx = nopass[0] 
        for i in nopass: 
            if dis[i] < dis[idx]: 
                idx = i 
        nopass.remove(idx) 
        passed.append(idx) 
        for i in nopass: 
            if dis[idx] + mgraph[idx][i] < dis[i]: 
                dis[i] = dis[idx] + mgraph[idx][i] 
                dict_[str(i)] = dict_[str(idx)] + [idx] 
    return dis,dict_ 
dis,dict_ = startwith(0, mgraph)
print(dict_) 
s=dict_['10'] 
s1=[i+1 for i in s] 
XXX=np.zeros(len(s1)) 
YYY=np.zeros(len(s1)) 
for i in range(len(s1)): 
    PP=eval('p'+str(s1[i])) 
    XXX[i]=PP[0] 
    YYY[i]=PP[1] 
XXX=np.insert(XXX,0,[0]) 
XXX=np.append(XXX,p11[0]) 
YYY=np.insert(YYY,0,[0]) 
YYY=np.append(YYY,p11[1]) 
plt.plot(XXX,YYY)
plt.show()
```
