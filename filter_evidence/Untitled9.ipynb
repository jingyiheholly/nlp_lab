```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming `length` is already populated as a list of integers
# Example dummy data
length = [1, 2, 3, 3, 4, 2, 1, 5, 4, 3, 3, 2, 6, 7, 8, 1, 0]  # Replace with actual list

# Convert to a numpy array for further analysis
length_array = np.array(length)

# Basic statistics
mean_length = np.mean(length_array)
median_length = np.median(length_array)
max_length = np.max(length_array)
min_length = np.min(length_array)
std_dev = np.std(length_array)

print(f"Mean: {mean_length}")
print(f"Median: {median_length}")
print(f"Max: {max_length}")
print(f"Min: {min_length}")
print(f"Standard Deviation: {std_dev}")

# Histogram
plt.figure(figsize=(12, 6))
plt.hist(length, bins=10, color='blue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Relation Lengths', fontsize=16)
plt.xlabel('Number of Relations', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Boxplot
plt.figure(figsize=(8, 4))
plt.boxplot(length, vert=False, patch_artist=True, boxprops=dict(facecolor='orange', color='black'))
plt.title('Boxplot of Relation Lengths', fontsize=16)
plt.xlabel('Number of Relations', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Violin-style plot (approximation with KDE curve and boxplot overlay)
plt.figure(figsize=(12, 6))
count, bins, _ = plt.hist(length, bins=10, density=True, alpha=0.5, color='green', edgecolor='black')
center = (bins[:-1] + bins[1:]) / 2
plt.plot(center, count, color='green', label='Density Curve', linewidth=2)
plt.boxplot(length, vert=False, positions=[0.05], widths=0.01, patch_artist=True, boxprops=dict(facecolor='red', color='black'))
plt.title('Violin-style Plot of Relation Lengths', fontsize=16)
plt.xlabel('Number of Relations', fontsize=12)
plt.legend()
plt.show()
```

```python
import torch

if torch.cuda.is_available():
    current_device = torch.cuda.current_device()
    total_memory = torch.cuda.get_device_properties(current_device).total_memory
    allocated_memory = torch.cuda.memory_allocated(current_device)
    free_memory = total_memory - allocated_memory

    print(f"Total GPU Memory: {total_memory / 1024 ** 3:.2f} GB")
    print(f"Allocated GPU Memory: {allocated_memory / 1024 ** 3:.2f} GB")
    print(f"Free GPU Memory: {free_memory / 1024 ** 3:.2f} GB")
else:
    print("CUDA is not available.")
```
