```python
from rdflib import Graph

g = Graph()
g.parse("http://dbpedia.org/sparql")  # Load RDF data

query = """
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?x 
WHERE { 
    ?x rdfs:label ?name . 
    FILTER(CONTAINS(LCASE(STR(?name)), "adam_mcquaid"))
}
"""

results = g.query(query)
for row in results:
    print(row)
```
